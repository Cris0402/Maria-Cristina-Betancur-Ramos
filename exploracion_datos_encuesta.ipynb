```python
# Obtengo los registros de los Jefes de hogar que tienen entre 25 y 42 años
datosset = datos[(datos['parentesco_jefe']=='Jefe') & (datos["edad"]<=42)& (datos["edad"]>=25)]
datosset
```

```python
#Creo nueva columna con valores de ingresos en dolares a valores del 2019, cotizacion del dolar ARS 63.00
ingresos_USD = {'ingresos_totales_USD': datosset['ingresos_totales']/63}
USD = pd.DataFrame(ingresos_USD)
datosset= pd.concat([datosset, USD], axis=1)
datosset
```

```python
# Con la funcion round formateamos la columna ingresos_totales_USD con 2 decimales
ingresos_totales_USD=round(datosset['ingresos_totales_USD'],2)
datosset
```

```python
#Obtengo la media de los ingresos totales
mediasalario= round(datos['ingresos_totales'].mean(),2)
mediasalario
```

```python
mediasalariodolares= mediasalario/63 #media salario en dolares año 2019
mediasalariodolares
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8), sharey=True)
mujeres = sns.boxplot(x='edad', y= datosset[datosset['sexo'] == 'Mujer']["ingresos_totales_USD"], data=datosset ,color="pink", ax=axes[0])
mujeres.axhline(y = mediasalariodolares, color="red")  #agrego el valor medio de ingreso como una linea roja
varones= sns.boxplot(x='edad', y = datosset[datosset['sexo'] == 'Varon']["ingresos_totales_USD"],data=datosset , color="skyblue", ax=axes[1])
varones.axhline(y = mediasalariodolares, color="red") #agrego el valor medio de ingreso como una linea roja
plt.show();
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 8),  sharey=True)
mujeres = sns.boxplot(y= datosset[(datosset['sexo'] == 'Mujer') & (datos["edad"]==35)]["ingresos_totales_USD"], data=datosset ,color="pink", ax=axes[0])
mujeres.axhline(y = mediasalariodolares, color="red")
varones= sns.boxplot(y = datosset[(datosset['sexo'] == 'Varon')& (datos["edad"]==35)]["ingresos_totales_USD"],data=datosset , color="skyblue", ax=axes[1])
varones.axhline(y = mediasalariodolares, color="red")
plt.show();
```
